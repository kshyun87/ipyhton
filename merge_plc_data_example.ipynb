## PLC 정보 파일(csv)불러오기
이때 PLC 데이터와 결합을 위해서 인덱스로 이큅먼트와 어드레스를 사용했습니다.

```python
data_plc_info = pd.DataFrame.from_csv("./origin/PLC_ITEM_INFO.csv", index_col=["EQUIPMENT_ID","ADDRESS"])
```

```python
data_plc_info[["SYSETM_DESCRIPTION", "EQUIPMENT_DESC","DIV_VALUE"]]
```

## 데이터 결합하기(join, merge)
이제 위 두개의 데이터 프레임을 합치는 작업입니다.
pandas의 concat함수로 두 데이터 프래임을 결합하여 하나의 데이터프레임으로 만들어줍니다.
이때 너무 많은데이터를 한번에 둘 필요가 없어서 PLC데이터중 시간, 값, 퀄리티만 남기고 결합하였습니다.

```python
result = pd.concat([data_frame[['time','value','quality']],data_plc_info[['SYSETM_DESCRIPTION']]],axis=1, join_axes=[data_frame[['time','value','quality']].index])
```

```python
result
```

### 데이터 인덱스를 컬럼으로 변경
에러코드를 분리하는데 편리하게 하기 위해서
위의 데이터 프레임에서 인덱스에 있던 이큅먼트와 어드레스를 컬럼으로 변경

## 데이터에서 에러코드를 추가하는 코드

```python
result_rest = result.reset_index(level=['equipment_id','address'])
result_rest['ERR']=''
result_rest['GOAL']=''
result_rest['PRODUCT']=''
result_rest['DEFECT']=''

len_result_rest = len(result_rest['equipment_id'])
# result_rest
# c = []
for each in range(0,len_result_rest):
    desc = result_rest.get_value(each,'SYSETM_DESCRIPTION')
    value = result_rest.get_value(each,'value')
    if desc =="ALARM_CODE":
        result_rest.set_value(each, 'ERR', value)
    else:
        result_rest.set_value(each, 'ERR', '')
    
    if desc =="GOAL_QTY":
        result_rest.set_value(each, 'GOAL', value)
    else:
        result_rest.set_value(each, 'GOAL', '')
        
    if desc =="PRODUCT_QTY":
        result_rest.set_value(each, 'PRODUCT', value)
    else:
        result_rest.set_value(each, 'PRODUCT', '')
        
    if desc =="DEFECT_QTY":
        result_rest.set_value(each, 'DEFECT', value)
    else:
        result_rest.set_value(each, 'DEFECT', '')
        
```

# 데이터에 에러코드 컬럼을 추가하여 만든 표

```python
result_rest[['time','equipment_id','address','value', 'ERR', 'GOAL', 'PRODUCT', 'DEFECT']]
```

## Index로 Time을 설정
시간으로 데이터를 검색하기 위해서 time컬럼을 Index로 변경한다.

```python
#time 컬럼값을 시간데이터로 변경(기존 문자열)
result_rest['time'] = pd.to_datetime(result_rest['time'])

#time 컬럼을 index로 설정.
fi_data = result_rest.set_index(['time'])

#결과 출력
fi_data
```

```python
# 날짜로 데이터 검색.
# 1월 7일 데이터만 검색
fi_data.ix["2016-01-07":"2016-01-07"]# ix[시작: 종료]로 날짜만 입력할 경우 날짜단위만 출력
```

```python
# ix[시작: 종료]로 시간까지 입력할 경우 날짜+시간단위로 검색하여 출력
fi_data.ix["2016-01-07 00:00:00":"2016-01-07 00:00:01"]
```
